## Descripción
Este notebook se conecta a MongoDB para leer los datos crudos almacenados en la
colección `raw_data` (base de datos `taller4_db`), los carga en un DataFrame de
pandas y realiza un análisis exploratorio: inspección, cálculo de insights y
visualizaciones.

```python
# Importación de librerías
import pandas as pd
from pymongo import MongoClient

# Conexión a MongoDB y lectura de la colección raw_data
cliente = MongoClient("mongodb://localhost:27017")
coleccion = cliente["taller4_db"]["raw_data"]

# Convertimos todos los documentos de MongoDB en un DataFrame de pandas
df = pd.DataFrame(list(coleccion.find()))

# Verificamos cuántos registros y columnas se cargaron
print(f"Registros cargados: {df.shape[0]}")
print(f"Columnas disponibles: {df.shape[1]}")
```

## 1. Selección de Variables

De las columnas disponibles se seleccionan 5 variables relevantes para el análisis:

- **title**: nombre del producto
- **category**: categoría del producto (variable categórica)
- **price**: precio en dólares
- **rating**: calificación promedio del producto
- **stock**: unidades disponibles en inventario

```python
# Seleccionamos únicamente las 5 columnas relevantes para el análisis
columnas = ["title", "category", "price", "rating", "stock"]
df = df[columnas]

# Confirmamos la nueva forma del DataFrame
print(f"DataFrame reducido a {df.shape[1]} columnas: {list(df.columns)}")
```

## 2. Análisis Exploratorio (EDA)

### A. Inspección Básica
Revisamos las primeras filas, los tipos de datos y la presencia de valores nulos.

```python
# Primeras 5 filas del DataFrame
df.head()
```

```python
# Información general: tipos de datos y memoria
df.info()
```

```python
# Verificamos cuántos valores nulos hay por columna
df.isnull().sum()
```

### B. Insights Numéricos (5 Datos Relevantes)
A continuación se calculan 5 métricas estadísticas que aportan valor al análisis del catálogo de productos.

```python
# Insight 1: Precio promedio de los productos
precio_promedio = df["price"].mean()
print(f"El precio promedio de los productos es ${precio_promedio:.2f}")
```

```python
# Insight 2: Producto más caro y más barato
mas_caro = df.loc[df["price"].idxmax()]
mas_barato = df.loc[df["price"].idxmin()]

print(f"Producto más caro: {mas_caro['title']} (${mas_caro['price']})")
print(f"Producto más barato: {mas_barato['title']} (${mas_barato['price']})")
```

```python
# Insight 3: Número de categorías y la categoría con más productos
num_categorias = df["category"].nunique()
categoria_top = df["category"].value_counts().idxmax()
cantidad_top = df["category"].value_counts().max()

print(f"Existen {num_categorias} categorías distintas de productos")
print(f"La categoría con más productos es '{categoria_top}' con {cantidad_top} productos")
```

```python
# Insight 4: Rating promedio y producto mejor calificado
rating_promedio = df["rating"].mean()
mejor_valorado = df.loc[df["rating"].idxmax()]

print(f"El rating promedio de los productos es {rating_promedio:.2f} de 5")
print(f"El producto mejor calificado es '{mejor_valorado['title']}' con rating {mejor_valorado['rating']}")
```

```python
# Insight 5: Stock total del inventario y producto con mayor stock
stock_total = df["stock"].sum()
mayor_stock = df.loc[df["stock"].idxmax()]

print(f"El inventario total suma {stock_total} unidades")
print(f"El producto con más stock es '{mayor_stock['title']}' con {mayor_stock['stock']} unidades")
```

### C. Visualización

#### Gráfico 1: Distribución de productos por categoría (Top 6)
Dado que existen 24 categorías, se muestran las 6 con más productos y el resto se agrupa en "Otras" para mantener la legibilidad del gráfico.

```python
import matplotlib.pyplot as plt

# Contamos productos por categoría y tomamos las 8 principales
conteo = df["category"].value_counts()
top8 = conteo.head(8)

# Agrupamos el resto de categorías en "Otras"
otras = conteo[8:].sum()
top8["Otras"] = otras

# Creamos el gráfico de torta
plt.figure(figsize=(8, 8))
plt.pie(top8, labels=top8.index, autopct="%1.1f%%", startangle=90)
plt.title("Distribución de productos por categoría (Top 8 + Otras)")
plt.axis("equal")  # Hace que la torta sea un círculo perfecto
plt.show()
```

#### Gráfico 2: Top 10 categorías por cantidad de productos
Gráfico de barras que compara las 10 categorías con mayor número de productos en el catálogo.

```python
# Top 10 categorías por cantidad de productos
top10 = df["category"].value_counts().head(10)

plt.figure(figsize=(10, 6))
top10.plot(kind="bar", color="steelblue")
plt.title("Top 10 categorías por cantidad de productos")
plt.xlabel("Categoría")
plt.ylabel("Cantidad de productos")
plt.xticks(rotation=45, ha="right")  # Rota las etiquetas para que se lean bien
plt.tight_layout()  # Ajusta el espaciado para que no se corten las etiquetas
plt.show()
```

#### Gráfico 3: Distribución de las calificaciones (rating)
Histograma que muestra cómo se distribuyen las calificaciones de los productos del catálogo.

```python
# Histograma de la distribución de ratings
plt.figure(figsize=(10, 6))
plt.hist(df["rating"], bins=10, color="mediumseagreen", edgecolor="black")
plt.title("Distribución de las calificaciones de los productos")
plt.xlabel("Rating (0 a 5)")
plt.ylabel("Cantidad de productos")
plt.show()
```
